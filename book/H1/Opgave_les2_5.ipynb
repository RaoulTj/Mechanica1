De functie voor de elastische lijn (doorbuigings) kan gevonden worden door gebruik te maken van;

$EI \dfrac{d^4v}{dx^4}= -q$ en deze twee keer te integreren.


\begin{align}
EI\dfrac{d^4v}{dx^4} &= -q \\
EI\dfrac{d^3v}{dx^3} &= -qx + C_1 \\
EI\dfrac{d^2v}{dx^2} &= -\dfrac{1}{2}qx^2 + C_1x + C_2 \\
EI\dfrac{dv}{dx} &= -\dfrac{1}{6}qx^3 +\dfrac{1}{2}C_1x^2 + C_2x + C_3 \\
EIv &= -\dfrac{1}{24}qx^4 +\dfrac{1}{6}C_1x^3 + \dfrac{1}{2}C_2x^2 + C_3x +C_4
\end{align}


De integratie constanten $C_1$, $C_2$, $C_3$ en $C_4$ kunnen gevonden worden door gebruik te maken van de volgende randvoorwaardes.

Bij oplegging $A$ geldt;

$x=0, v=0$


\begin{align}
    EIv &= -\dfrac{1}{24}qx^4 +\dfrac{1}{6}C_1x^3 + \dfrac{1}{2}C_2x^2 + C_3x +C_4 \\
    EI\cdot 0 &= -\dfrac{1}{24}q \cdot 0^4 +\dfrac{1}{6}C_1 \cdot 0^3 + \dfrac{1}{2}C_2 \cdot 0^2 + C_3 \cdot 0 +C_4  \\
    0 &= -0 +0 + 0 +0 + C_4 \\
    C_4 &= 0
\end{align}


Bij oplegging $A$ geldt;

$x=0, \dfrac{d^2v}{dx^2}=0$

\begin{align}
    EI\dfrac{d^2v}{dx^2} &= -\dfrac{1}{2}qx^2 + C_1x + C_2 \\
    EI \cdot 0 &= -\dfrac{1}{2}q \cdot 0^2 + C_1 \cdot 0 + C_2 \\
    EI \cdot 0 &= 0 + 0 + C_2\\
    C_2 &= 0
\end{align}

Bij oplegging $B$ geldt;

$x=l, \dfrac{d^2v}{dx^2}=0$


\begin{align}
    EI\dfrac{d^2v}{dx^2} &= -\dfrac{1}{2}qx^2 + C_1x + C_2 \\
    EI \cdot 0 &= -\dfrac{1}{2}q \cdot l^2 + C_1 \cdot l + 0 \\
    C_1 &= \dfrac{1}{2}q \cdot l
\end{align}


Bij oplegging $B$ geldt;

$x=l, v=0$


\begin{align}
    EIv &= -\dfrac{1}{24}qx^4 +\dfrac{1}{6}C_1x^3 + \dfrac{1}{2}C_2x^2 + C_3x +C_4 \\
    EI\cdot 0 &= -\dfrac{1}{24}q \cdot l^4 +\dfrac{1}{6}C_1 \cdot l^3 + \dfrac{1}{2} \cdot 0 \cdot l^2 + C_3 \cdot l + 0  \\
    EI\cdot 0 &= -\dfrac{1}{24}q \cdot l^4 +\dfrac{1}{6}C_1 \cdot l^3 + C_3 \cdot l  \\
    EI\cdot 0 &= -\dfrac{1}{24}q \cdot l^4 +\dfrac{1}{6}(\dfrac{1}{2}q \cdot l) \cdot l^3 + C_3 \cdot l  \\
    EI\cdot 0 &= -\dfrac{1}{24}q \cdot l^4 +\dfrac{1}{12}q \cdot l^4 + C_3 \cdot l  \\
    EI\cdot 0 &= \dfrac{1}{24}q \cdot l^4  + C_3 \cdot l  \\
    C_3 &= -\dfrac{1}{24}q \cdot l^3
\end{align}

De waardes voor $C_1$, $C_2$, $C_3$ en $C_4$ invullen geeft:


\begin{align}
EI\dfrac{d^4v}{dx^4} &= -q \\
EI\dfrac{d^3v}{dx^3} &= -qx + \dfrac{1}{2}ql \\
EI\dfrac{d^2v}{dx^2} &= -\dfrac{1}{2}qx^2 + \dfrac{1}{2}qlx \\
EI\dfrac{dv}{dx} &= -\dfrac{1}{6}qx^3 +\dfrac{1}{4}qlx^2  -\dfrac{1}{24}ql^3 \\
EIv &= -\dfrac{1}{24}qx^4 +\dfrac{1}{12}qlx^3 -\dfrac{1}{24}q \cdot l^3x
\end{align}


De functie voor de dwarskracht is bekend en kan nu getekend worden;

\begin{align}
    V = EI\dfrac{d^3v}{dx^3} &= -qx + \dfrac{1}{2}ql \\
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 20     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
V_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        V_values[i] = -q * x + (4 / 8) *q * L
    else:  # Rechts van het koppel
        V_values[i] = -q * x + (4 / 8) *q * L  # Sprong in moment door koppel

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, V_values, label="Dwarskrachtverloop V(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk (m)")
plt.ylabel("Drarskracht V(x) [kN]")
plt.title("Dwarskrachtenlijn van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

$A_v$ kan gevonden worden door $x=0$ in te vullen in de functie voor de dwarskracht;

\begin{align}
    A_v = EI\dfrac{d^3v}{dx^3} &= -qx + \dfrac{1}{2}ql \\
    A_v &= -q \cdot 0 + \dfrac{1}{2}ql \\
    A_v &= \dfrac{1}{2}ql \\
    A_v &= \dfrac{1}{2} \cdot 10 \cdot 10 = 50 kN \\
\end{align}

De functie voor het buigendmoment is bekend en kan nu getekend worden;

\begin{align}
    M = EI\dfrac{d^2v}{dx^2} &= -\dfrac{1}{2}qx^2 + \dfrac{1}{2}qlx
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 11     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
M_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        M_values[i] = - (q / 2) * x**2 + (4 / 8) * q * L * x
    else:  # Rechts van het koppel
        0

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, M_values, label="Momentverloop M(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk (m)")
plt.ylabel("Moment M(x) [kNm]")
plt.title("Momentenlijn van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

\begin{align}
    M_{max} = EI\dfrac{d^2v}{dx^2} &= -\dfrac{1}{2}q \cdot (\dfrac{1}{2}l)^2 + \dfrac{1}{2}ql \cdot \dfrac{1}{2}l \\

De functie van de helling is bekend en kan nu getekend worden;


\begin{align}
    EI\dfrac{dv}{dx} &= -\dfrac{1}{6}qx^3 +\dfrac{1}{4}qlx^2  -\dfrac{1}{24}ql^3 
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 11     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
EIv_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        EIv_values[i] = (-(1/6) * q * x**3 + (1/4) * q * L * x**2 - (1/24) * q * L**3 ) * 1
    else:  # Rechts van het koppel
        EIv_values[i] = 0

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, EIv_values, label=r"$EI\dfrac{dv(x)}{dx}$", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk [m]")
plt.ylabel(r"$EI\dfrac{dv(x)}{dx}$ [m]")
plt.title("Helling van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

De functie van de elastischelijn (doorbuigingslijn) is bekend en kan nu getekend worden;


\begin{align}
E   Iv &= -\dfrac{1}{24}qx^4 +\dfrac{1}{12}qlx^3 -\dfrac{1}{24}q \cdot l^3x
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 11     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
EIv_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        EIv_values[i] = (-(1/24) * q * x**4 + (1/12) * q * L * x**3 - (1/24) * q * L**3 * x) * 1
    else:  # Rechts van het koppel
        EIv_values[i] = 0

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, EIv_values, label=r"$EIv$", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk [m]")
plt.ylabel(r"$EIv$ [m]")
plt.title("Doorbuiging van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
